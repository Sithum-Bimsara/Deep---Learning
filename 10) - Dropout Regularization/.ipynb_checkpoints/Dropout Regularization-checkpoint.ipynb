# Dropout Regularization

Dropout is a regularization technique used to prevent overfitting in neural networks by randomly "dropping out" units (along with their connections) during training.

---

## Key Points

1. **Purpose**:
   - Reduces overfitting by making the network less reliant on specific neurons.
   - Encourages the network to learn more robust and distributed representations.

2. **How It Works**:
   - During each training iteration, a random subset of neurons is "dropped out" by setting their activations to zero.
   - Dropout is applied only during training and not during inference.

3. **Dropout Rate**:
   - Specifies the fraction of neurons to be dropped.
   - Common values: `0.2` (20%) to `0.5` (50%).

4. **Impact**:
   - Acts as a form of model averaging by effectively training different sub-networks during each iteration.
   - Reduces co-dependency among neurons.

---

````python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
````

````python
# load dataset
data = pd.read_csv("sonar.csv", header=None)
data = data.values
````

````python
data[:1]
````

````python
# divide dataset into x and y
X = data[:,0:60].astype(float)
Y = data[:,60]
````

````python
# encode labels
encoder = LabelEncoder()
encoder.fit(Y)
Y = encoder.transform(Y)
````

````python
# scale features
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
````

````python
# divide dataset into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
````

````python
tf.random.set_seed(42)

# # create a model without dropout layers
# model = tf.keras.Sequential([
#     tf.keras.layers.Dense(30, activation='relu', input_shape=(60,)),
#     tf.keras.layers.Dense(60, activation='relu'),
#     tf.keras.layers.Dense(60, activation='relu'),
#     tf.keras.layers.Dense(30, activation='relu'),
#     tf.keras.layers.Dense(1, activation='sigmoid')
# ])

# create a model with droput layers
model = tf.keras.Sequential([
    tf.keras.layers.Dense(30, activation='relu', input_shape=(60,)),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(60, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(60, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(30, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
````

````python
# compile the model
model.compile(loss = tf.keras.losses.BinaryCrossentropy(),
             optimizer = tf.keras.optimizers.Adam(learning_rate=0.005),
             metrics = ["accuracy"])
````

````python
# train the model
epoch_number = 100
history = model.fit(X_train, y_train, epochs=epoch_number)
````

````python
# evaluate the model
model.evaluate(X_test, y_test)
````
